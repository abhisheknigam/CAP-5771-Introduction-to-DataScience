```python
import matplotlib.pyplot as pp
import collections
% matplotlib inline
```

```python
years = [1950, 1960, 1970, 1980, 1990, 2000, 2010]
gdp = [300.2, 543.3, 1075.9, 2862.5, 5979.6, 10289.7, 14958.3]
```

```python
pp.plot(years,gdp, color = 'green', marker = 'o', linestyle='solid')
pp.title("GDP")
pp.ylabel("Billions of $")
```

```python
##########
######BAR CHARTS
##########
movies = ["Annie Hall", "Ben-Hur", "Casablanca", "Gandhi", "West Side Story"]
num_oscars = [5, 11, 3, 8, 10]
```

```python
xs = xs = [i + .1 for i, _ in enumerate(movies)]
pp.bar(xs, num_oscars)
pp.ylabel("# of Academy Awards")
pp.title("My Favorite Movies")
pp.xticks([i + 0.5 for i, _ in enumerate(movies)], movies)
```

```python
grades = [83,95,91,87,70,0,85,82,100,67,73,77,0]
decile = lambda grade: grade // 10 * 10
histogram = collections.Counter(decile(grade) for grade in grades)
pp.bar([x - 4 for x in histogram.keys()], # shift each bar to the left by 4
histogram.values(), # give each bar its correct height
8)
pp.axis([-5,105,0,5])
pp.xticks([10 * i for i in range(11)]) # x-axis labels at 0, 10, ..., 100
pp.xlabel("Decile")
pp.ylabel("# of Students")
pp.title("Distribution of Exam 1 Grades")
```
